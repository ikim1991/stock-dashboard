```python
balance_sheets.columns = ['ticker', 'quarter_end', 'currency', 'total_assets', 'total_liabilities', 'intangible_assets', 'shareholders_equity', 'total_equity']
balance_sheets['notice'] = 'All values in millions'
```

```python
for i in range(3,8):
    for j in range(0, len(balance_sheets)):
        if type(balance_sheets.iloc[j,i]) != str:
            continue
        elif 'B' in balance_sheets.iloc[j,i] and '(' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('B', '').strip('(').strip(')')) * -1000000000
        elif 'B' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('B', '')) * 1000000000
        elif 'M' in balance_sheets.iloc[j,i] and '(' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('M', '').strip('(').strip(')')) * -1000000
        elif 'M' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('M', '')) * 1000000
        elif 'T' in balance_sheets.iloc[j,i] and '(' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('T', '').strip('(').strip(')')) * -1000000
        elif 'T' in balance_sheets.iloc[j,i]:
            balance_sheets.iloc[j,i] = float(balance_sheets.iloc[j,i].replace('T', '')) * 1000000
        else:
            balance_sheets.iloc[j,i] = 0
            
```

```python
balance_sheets['exchange_rate'] = np.nan
for i,v in enumerate(balance_sheets.currency):
    if v == 'USD':
        balance_sheets.exchange_rate[i] = 1.0
    elif type(v) == str:
        r = requests.get('https://www.exchange-rates.org/converter/{}/USD/1'.format(v))
        soup = bs4.BeautifulSoup(r.text, "html.parser")
        exchange_rate = float(soup.findAll('span',{'id':'ctl00_M_lblToAmount'})[0].text.strip())
        balance_sheets.exchange_rate[i] = exchange_rate
    else:
        continue
```

```python
company_listing = pd.read_csv('./datasets/companylisting.csv')
company_listing.drop(['Unnamed: 0'], inplace=True, axis=1)
company_listing.head()
```

```python
company_listing.sector.value_counts()
```

```python
gics_codes = {
    'Industrials': 'Industrials',
    'Financials': 'Financials',
    'Financial Services': 'Financials',
    'Consumer Discretionary': 'Consumer Discretionary',
    'Consumer Cyclical': 'Consumer Discretionary',
    'Health Care': 'Health Care',
    'Healthcare': 'Health Care',
    'Energy': 'Energy',
    'Utilities': 'Utilities',
    'Real Estate': 'Real Estate',
    'Consumer Staples': 'Consumer Staples',
    'Consumer Defensive': 'Consumer Staples',
    'Information Technology': 'Information Technology',
    'Technology': 'Information Technology',
    'Materials': 'Materials',
    'Basic Materials': 'Materials',
    'Communication Services': 'Communication Services'
}
```

```python
company_listing.sector = company_listing.sector.map(gics_codes)
```

```python
company_listing.sector.value_counts()
```

```python
df = company_listing
```

```python
## Connect to SQL database
conn = psycopg2.connect(database = 'cuzegotk',user = 'cuzegotk',password = 'NekW2BqJ8hW1wO3hCdpuEESPiP-y131V',host = 'raja.db.elephantsql.com')
cur = conn.cursor()
## Transfer CSV to SQL Database
for i in range(0,len(df)):
    cur.execute('UPDATE companylisting SET sector = %s WHERE ticker = %s' , (str(df.iloc[i,-1]), str(df.iloc[i,0])))
    conn.commit()
cur.close()
```
